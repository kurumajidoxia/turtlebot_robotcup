```python
import rospy
import math
from sensor_msgs.msg import Image
from aruco_msgs.msg import MarkerArray
from cv_bridge import CvBridge, CvBridgeError

bridge = CvBridge()
```

```python
rospy.init_node('turtlebot_markers',anonymous=True)
rospy.sleep(0.5)
```

```python
def callback_image(msg):
    global marker_img
    global marker_timestamp
    marker_img_timestamp = msg.header.stamp
    marker_img = bridge.imgmsg_to_cv2(msg, "rgb8")

def callback_array(msg):
    global marker_array
    global marker_array_timestamp
    marker_array_timestamp = msg.header.stamp
    marker_array = msg.markers
```

```python
subscriber_img = rospy.Subscriber("/bestRobot/aruco_marker_publisher/result", Image, callback_image)
subscriber_array = rospy.Subscriber("/bestRobot/aruco_marker_publisher/markers", MarkerArray, callback_array)
```
